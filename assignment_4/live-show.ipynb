```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
import requests
import csv

header = ['name','address','url']
data = []

url = "https://www.showstart.com/venue/list?cityCode=10&t=1633932428226"
html = urlopen(url)
soup = BeautifulSoup(html,'html.parser')
for item in soup.find_all("div", attrs={"class":"table-wrap-cell_body"}):
    name = item.find("a",attrs={"class":"name"}).get_text().strip()
    address = item.find("div",attrs={"class":"addr"})
    url = item.find("a",attrs={"class":"num"}).get('href')
    data.append((name, address, url))
        
df = pd.DataFrame(data,columns = header)
df.to_csv('liveshow.csv', sep='\t', encoding='utf-8') 
df
```
